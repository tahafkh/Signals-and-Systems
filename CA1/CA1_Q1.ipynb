Question 1: Plotting different systems

# 1.1

First we define $x[n]$ to use it for plotting different functions:

$x[n] = Acos[2\pi kn]$

```python
import numpy as np
from matplotlib import pyplot as plt

A = 20
k = 0.6

# T = (2 * pi)/(2 * pi * k) = 1 / k
T = 1/k

title = "Figure of function {}"
def x(n):
    return A*np.cos(2*np.pi*k*n)

def draw(name, x_, y_):
    plt.stem(x_, y_, use_line_collection=True)
    plt.title(title.format(name))
    plt.show()
x_samples = np.linspace(0, 20, 200)
y_samples = x(x_samples)
draw("x[n]", x_samples, y_samples)
```

$A: y[n] = \frac{1}{T}(x[n+1]-x[n])$

```python
def y1(n):
    return (x(n+1) - x(n))/T

y_samples = y1(x_samples)
draw("A", x_samples, y_samples)
```

$B: y[n] = \frac{1}{2T}(x[n+1]-x[n])$

```python
def y2(n):
    return (x(n+1) - x(n))/2*T
y_samples = y2(x_samples)
draw("B", x_samples, y_samples)
```

$C: y[n] = \frac{1}{T^2}(x[n+1]-2x[n] + x[n-1])$

```python
def y3(n):
    return (x(n+1) - 2*x(n) + x(n-1))/T*T
y_samples = y3(x_samples)
draw("C", x_samples, y_samples)
```

$D: y[n] = 0.1x[n-2] + 0.2x[n-1] + 0.4x[n] + 0.2x[n+1] + 0.1x[n+2]$

```python
def y4(n):
    return 0.1*x(n-2) + 0.2*x(n-1) + 0.4*x(n) + 0.2*x(n+1) + 0.1*x(n+2)
y_samples = y4(x_samples)
draw("D", x_samples, y_samples)
```
