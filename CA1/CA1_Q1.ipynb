# 1.1

First we define $$x[n]$$ to use it for plotting different functions:

$$x[n] = Acos[2\pi kn]$$

```python
import numpy as np
from matplotlib import pyplot as plt

A = 20
k = 0.05

# Sampling period
T = 1

title = "Figure of function {}"
def x(n):
    return A*np.cos(2*np.pi*k*n)

def draw(name, x_, y_):
    plt.stem(x_, y_, use_line_collection=True)
    plt.title(title.format(name))
    plt.show()
x_samples = np.arange(0, 40, 2)
y_samples = x(x_samples)
draw("x[n]", x_samples, y_samples)
```

$$A: y[n] = \frac{1}{T}(x[n+1]-x[n])$$

```python
systems = list()
systems.append(lambda n: (x(n+1) - x(n))/T)

y_samples = systems[0](x_samples)
draw("A", x_samples, y_samples)
```

$$B: y[n] = \frac{1}{2T}(x[n+1]-x[n])$$

```python
systems.append(lambda n: (x(n+1) - x(n))/(2*T))

y_samples = systems[1](x_samples)
draw("B", x_samples, y_samples)
```

$$C: y[n] = \frac{1}{T^2}(x[n+1]-2x[n] + x[n-1])$$

```python
systems.append(lambda n: (x(n+1) - 2*x(n) + x(n-1))/(T**2))

y_samples = systems[2](x_samples)
draw("C", x_samples, y_samples)
```

$$D: y[n] = 0.1x[n-2] + 0.2x[n-1] + 0.4x[n] + 0.2x[n+1] + 0.1x[n+2]$$

```python
systems.append(lambda n: 0.1*x(n-2) + 0.2*x(n-1) + 0.4*x(n) + 0.2*x(n+1) + 0.1*x(n+2))

y_samples = systems[3](x_samples)
draw("D", x_samples, y_samples)
```

```python
x_samples = np.arange(0, 40, 1)
y_samples = x(x_samples)
draw("x[n] with increased samples", x_samples, y_samples)
```

# Conclusion
As you can see, for function A, y[n] is similar to -sin, which is first derivative of x[n] (cos) and it's reasonable because it's the difference of x[n+1] and x[n].
Function B is similar to function A(just different in coefficient). In function C, since we have double difference in system, it's similar to second deriative
and as a result, it becomes -cos which is second derivative of cos. Finally, in function D, since y[n] is a linear combination of x[n] and it's shifted functions
and since the sum of coefficients is 1, y[n] becomes similar to x[n] (since the sampling period is close enough to capture most input),
so y[n] becomes similar to x[n] (cos) as it could be seen from figures.
